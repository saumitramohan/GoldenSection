```python
import math
import numpy as np
from pandas import DataFrame
from math import e
import time

def funcOne(x):
    return pow(e,x)-7*x
funcOne_optimum = 1.9459101491

def funcTwo(x):
    return 4*x*x+5*x+2
funcTwo_optimum=-0.625

e_r = np.sqrt(1e-15)
e_f = 1e-6
e_abs = 1e-4
GSTestResult = DataFrame(columns=['f(x)', 'delta','Numoffunceval', 'Running time(ms)','d_rel']);

def GoldenSectionFunction(_func, delta):
    tou = (np.sqrt(5)-1)/2
    x1 = 10*np.random.randn(1)
    f1 = _func(x1)
    x2 = x1 + delta
    f2 = _func(x2)
    if f2>f1:
        x1,x2=x2,x1
        f1 = _func(x1)
        f2 = _func(x2)
        delta = -delta
    counter = 0
    while (True):
        delta = delta / tou
        x4 = x2 + delta
        f4 = _func(x4)
        if f4 > f2:
            break
        x1,x2=x2,x4
        f1 = _func(x1)
        f2 = _func(x2)
        counter= counter+1
        #check for extreme value
        if counter > 1e6:
            break
    _func_old = (f1+f2+f4)/3
    counter = 0
    while(True):
        x3 = tou * x4 + (1-tou)*x1
        f3 = _func(x3)
        if f2 < f3:
            x4,x1=x1,x3
            f4 = _func(x4)
            f1 = _func(x1)
        else:
            x1,x2=x2,x3
            f1 = _func(x1)
            f2 = _func(x2)
        _func_curr = (f1+f2+f4)/3
        if abs(_func_curr - _func_old ) <= e_r*abs(f2) + e_abs:
            break
        _func_old  = _func_curr
        if abs(x1-x4) <= e_r * abs(x2) + e_abs :
            break
        counter = counter + 1
        if counter > 1e6:
            break
    return x2[0],f2[0],counter

def _test_func (_func,mathOptimum,numberofRuns,delta):
    count = 1
    global GSTestResult;
    GSTable = DataFrame(columns=['Func','NumberofRuns','RunningTime','xOptimum','fOptimum','relDistance'])
    
    while (count <= numberofRuns):
        startTime = time.time()
        x_optValue,f_optValue,counter = GoldenSectionFunction (_func, delta)
        runTime = (time.time() - startTime)*pow(10,3)
        relDistance = abs(x_optValue - mathOptimum)/abs(x_optValue)
        count = count + 1
        GSTable.loc[len(GSTable)]=[_func.__name__,numberofRuns,runTime,x_optValue,f_optValue,relDistance]
        
    testResultFile=open('./GoldenSection/'+_func.__name__+'.txt','w+')
    testResultFile.write('Golden Section test Results\nAverage running time'+str(GSTable.RunningTime.mean())+\
                         '+-'+str(np.sqrt(1.0/(numberofRuns-1))*GSTable.RunningTime.std())+\
                         '\nrelDistance: '+str(GSTable.relDistance.mean())+\
                         ' +- '+str(np.sqrt(1.0/(numberofRuns-1))*GSTable.relDistance.std()) +'\n')  
    testResultFile.close()
        
    GSTestResult.loc[len(GSTestResult)]=[_func.__name__,delta,numberofRuns,GSTable.RunningTime.mean()\
                                         ,GSTable.relDistance.mean()]


if __name__ == '__main__':
    step = 0.3
    _test_func (funcOne, funcOne_optimum, 120, step)
    step= 1.0
    _test_func (funcOne, funcOne_optimum, 120, step)
    step= 2.0
    _test_func (funcOne, funcOne_optimum, 120, step)
    step = 0.3
    _test_func (funcTwo, funcTwo_optimum, 120, step)
    step= 1.0
    _test_func (funcTwo, funcTwo_optimum, 120, step)
    step= 2.0
    _test_func (funcTwo, funcTwo_optimum, 120, step)
    GSTestResult.to_csv('./GoldenSection/Result.csv')
    
    
            
            
        
    
















```
